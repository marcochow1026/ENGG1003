```python
#Q2b

import numpy as np

def log_likelihood(beta1, x, y, beta0):
    z = beta0 + beta1 * x
    # Use safe computation to avoid overflow
    return np.sum(y * z - np.log(1 + np.exp(z)))

def l_prime(beta1, x, y, beta0):
    z = beta0 + beta1 * x
    p = 1 / (1 + np.exp(-z))
    return np.sum(x * (y - p))

def l_double_prime(beta1, x, y, beta0):
    z = beta0 + beta1 * x
    p = 1 / (1 + np.exp(-z))
    return -np.sum(x**2 * p * (1 - p))

def newtons_method(x, y, beta0, cutoff, beta1_init=0.0):
    beta1 = beta1_init
    l_old = log_likelihood(beta1, x, y, beta0)
    beta_trace = [beta1]
    l_trace = [l_old]
    iteration = 0
    while True:
        iteration += 1
        lp = l_prime(beta1, x, y, beta0)
        lpp = l_double_prime(beta1, x, y, beta0)
        if lpp == 0:  # Avoid division by zero
            print("Second derivative is zero, stopping.")
            break
        beta1 = beta1 - lp / lpp
        l_new = log_likelihood(beta1, x, y, beta0)
        beta_trace.append(beta1)
        l_trace.append(l_new)
        if abs(l_new - l_old) <= cutoff:
            break
        l_old = l_new
    return beta_trace, l_trace

# Example usage (replace with loaded data)
#data = np.genfromtxt('HW1Q2data1.csv', delimiter=',', skip_header=1)
#x = data[:, 0]
#y = data[:, 1]
#beta_trace, l_trace = newtons_method(x, y, beta0=-0.66, cutoff=1e-14)
```
